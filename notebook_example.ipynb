```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import PathPatch
from matplotlib.path import Path
```

```python
N = 400
t = np.linspace(0, 2*np.pi, N)
r = 0.5 + np.cos(t)
x = r * np.cos(t)
y = r * np.sin(t)
```

```python
fing, ax = plt.subplots()
ax.plot(x, y, "k")
ax.set(aspect=1)
plt.show()
```
